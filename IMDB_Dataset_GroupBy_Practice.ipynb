```python
# find number of movies starting with A for each group
def movie_starts_with_A(group):
  return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(movie_starts_with_A)
```

```python
# find rating of each movie in the group according to IMDB score
# assigning the rank of movie using IMDB rating of its group
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB Rating group wise
#  ( (x - xmin) / (xmax - xmin) )
def normal(group):
  x = group['IMDB_Rating']
  group['Norm_Rating'] = ( (x - x.min()) / (x.max() - x.min()) )
  return group

genres.apply(normal)
```

```python
# Create a function that can return the highest score of any batsman
temp2 = ipl[ipl['batsman'] == 'V Kohli']
temp2.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest_score(batsman):
  temp2 = ipl[ipl['batsman'] == batsman]
  return temp2.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest_score('CH Gayle')
```
