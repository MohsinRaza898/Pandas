**dropna**
*   By default dropna(how='any') if nan in any row its removes these row
*   (how = 'all') know it removes row if missing values in complete row
*   (subset = ['name']) it removes row if missing value in perticular col, for multiple (subset = ['name1', 'name2']




```python
# dropna (drop missing values)
# series
ipl['method'].dropna()
```

```python
len(ipl['method'].dropna())
```

```python
# dataframe
ipl.dropna() # if 1 issing value in any row it remove row
```

**fillna**
*   .fillna(xyz) replace missing value with xyz(series)
*   .fillna(xyz) replace every missing value with xyz(dataframe) but always do col by col
*   data name[col name],fillna(value) value for missing value
*   .fillna(method='ffill' or 'bfill')
*   ffill(forward fill) filling previous value in next nan value
*   bfill(before fill) filling next value in previous nan value








```python
# fillna (fill missing values)
ipl['method'].fillna('unkown').hasnans
```

```python
# find the last match played by virat kohli in delhi
ipl['all_players'] = ipl['Team1Players'] + ipl['Team2Players']
ipl.head(2)
```

```python
def did_kohli_play(players_list):
  return 'V Kohli' in players_list
```

```python
ipl['did_kohli_play'] = ipl['all_players'].apply(did_kohli_play)
ipl[(ipl['City'] == 'Delhi') & (ipl['did_kohli_play'] == True)].drop_duplicates(subset=['City','did_kohli_play'],keep='first')
```
